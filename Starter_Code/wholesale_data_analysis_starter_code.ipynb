## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
clear_df = df.copy()
clear_df["line_subtotal"] = clear_df["unit_price"] * clear_df["qty"]
display_columns = ["unit_price", "qty", "line_subtotal"]
clear_df[display_columns].head(2)

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
clear_df = clear_df.assign(total_weight=df.unit_weight * df.qty)
clear_df = clear_df.assign(shipping_price = [weight * 7 if weight > 50 else weight * 10 for weight in clear_df["total_weight"]])

display_columns = ["unit_price", "unit_weight", "qty", "total_weight", "shipping_price"]
clear_df[display_columns].head(3)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
clear_df["line_price"] = ((clear_df.line_subtotal + clear_df.shipping_price) * 109.25 /100).round(2)
display_columns = ["line_subtotal", "shipping_price", "line_price"]

clear_df[display_columns].head(3)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
clear_df["line_cost"] = clear_df.unit_cost * clear_df.qty + clear_df.shipping_price
clear_df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
clear_df["line_profit"] = clear_df.line_price - clear_df.line_cost
clear_df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

for client in top_5_client_list:
    spend = clear_df.loc[clear_df.client_id == client]["line_price"].sum().round(2)
    print(f"{client}: &{spend}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

def get_top_client_info():
    display_columns = ["qty", "shipping_price", "line_price", "line_cost", "line_profit"]

    data = []
    for id in top_5_client_list:
        temp_df = clear_df.loc[clear_df.client_id == id]
        row = temp_df[display_columns].sum(axis=0)
        newRow = pd.concat([pd.Series({"client_id": id}), row])
        data.append(newRow)

    new_df = pd.DataFrame(data)
    new_df['client_id'] = new_df['client_id'].astype(int)
    new_df['qty'] = new_df['qty'].astype(int)
    
    return new_df

top_five_client_df = get_top_client_info()
top_five_client_df
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
money_column = ["shipping_price", "line_price", "line_cost", "line_profit"]

# Define a function that converts a dollar amount to millions.
def currency_format_millions(amount):
    return amount / 1_000_000

# Apply the currency_format_millions function to only the money columns. 
top_five_client_df[money_column] = top_five_client_df[money_column].apply(lambda x: currency_format_millions(x))

# Rename the columns to reflect the change in the money format. 
top_five_client_df = top_five_client_df.rename(columns={"client_id": "Client ID",
                                   	"qty": "Units",
                                    "shipping_price": "Shipping (millions)",
                                    "line_price": "Total Revenue (millions)",
                                    "line_cost": "Total Cost (millions)",
                                    "line_profit": "Total Profit (millions)"})

top_five_client_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
final_df = top_five_client_df.sort_values(by="Total Profit (millions)", ascending=False).copy()
final_df
```
